```python
%matplotlib inline
import scipy
import sympy
import numpy as np
import matplotlib.pyplot as plt
from modsim import*

condition = Condition(R = 5100,
                      L = 0.1,
                      C0 = 220 * (10 ** (-9)), 
                      V0 = 5,
                      dV0dt = 0,
                      f0 = 1073.02, 
                      w0 = 1)

# Do research of temp dependance of L and C
```

```python
duration = 60

dt = 1

ts = linrange(0, duration, dt)
```

```python
def make_system(condition):
    
    unpack(condition)
    
    init = State(X1 = V0, X2 = dV0dt)
    
    system = System(init=init, R=R, L=L, C0=C0, V0=V0, dV0dt=dV0dt, f0=f0, w0=w0, ts=ts)
    
    
    return system
                    
    
circuit = make_system(condition)
```

```python
def slope_func(state, t, circuit):
    
    X1, X2 = state
    unpack(circuit)
    
    dX1dt = X2
    dX2dt = ((L/R*C0) * (V0*np.degrees(np.sin(w0*t))) - ((1/R*L*C0)*X1) - ((1/L)*X2)) 

    
    return dX1dt, dX2dt

slope_func(circuit.init, 0, circuit)
```

```python
run_odeint(circuit, slope_func) # run a for loop around run_odeint

circuit.results
```
